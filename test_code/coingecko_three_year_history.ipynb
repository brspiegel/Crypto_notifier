```python
import pandas as pd
import datetime


```

```python
from pycoingecko import CoinGeckoAPI
cg = CoinGeckoAPI()
```

```python
# Create unix from date
start_date_three_year = pd.Timestamp('2019-09-01').timestamp()
start_date_three_year
```

```python
# Create unix end date
end_date_three_year = pd.Timestamp('2022-09-01').timestamp()
end_date_three_year
```

```python
# Create variables for cryptocurrency data pull
symbols = ['bitcoin', 'ethereum', 'ripple', 'cardano', 'solana']
```

```python
# pull bitcoin data for three years
bitcoin_three_year = cg.get_coin_market_chart_range_by_id(id=f'{symbols[0]}', vs_currency='usd', from_timestamp=start_date_three_year, to_timestamp=end_date_three_year)
bitcoin_three_year = bitcoin_three_year['prices']
bitcoin_three_year_prices = pd.DataFrame(bitcoin_three_year)
bitcoin_three_year_prices.columns = ['date', 'price']
bitcoin_three_year_prices['daily_return'] = bitcoin_three_year_prices['price'].pct_change().dropna()
bitcoin_three_year_prices.head()
```

```python
# pull etherium data for three years
ethereum_three_year = cg.get_coin_market_chart_range_by_id(id=f'{symbols[1]}', vs_currency='usd', from_timestamp=start_date_three_year, to_timestamp=end_date_three_year)
ethereum_three_year = ethereum_three_year['prices']
ethereum_three_year_prices = pd.DataFrame(ethereum_three_year)
ethereum_three_year_prices.columns = ['date', 'price']
ethereum_three_year_prices['daily_return'] = ethereum_three_year_prices['price'].pct_change().dropna()
ethereum_three_year_prices.head()
```

```python
# pull ripple data for three years
ripple_three_year = cg.get_coin_market_chart_range_by_id(id=f'{symbols[2]}', vs_currency='usd', from_timestamp=start_date_three_year, to_timestamp=end_date_three_year)
ripple_three_year = ripple_three_year['prices']
ripple_three_year_prices = pd.DataFrame(ripple_three_year)
ripple_three_year_prices.columns = ['date', 'price']
ripple_three_year_prices['daily_return'] = ripple_three_year_prices['price'].pct_change().dropna()
ripple_three_year_prices.head()
```

```python
# pull cardano data for three years
cardano_three_year = cg.get_coin_market_chart_range_by_id(id=f'{symbols[3]}', vs_currency='usd', from_timestamp=start_date_three_year, to_timestamp=end_date_three_year)
cardano_three_year = cardano_three_year['prices']
cardano_three_year_prices = pd.DataFrame(cardano_three_year)
cardano_three_year_prices.columns = ['date', 'price']
cardano_three_year_prices['daily_return'] = cardano_three_year_prices['price'].pct_change().dropna()
cardano_three_year_prices.head()
```

```python
# create a merge dictionary 
to_merge_dict = {'bitcoin': bitcoin_three_year_prices.loc[:, ['date','daily_return']] , 'ethereum': ethereum_three_year_prices.loc[:, 'daily_return'], 'cardano': cardano_three_year_prices.loc[:,'daily_return'], 'ripple': ripple_three_year_prices.loc[:, 'daily_return'], 'solana': solana_three_year_prices.loc[:,'daily_return']}
```

```python
# do the merge
merged_df = pd.concat(to_merge_dict.values(), axis=1, keys=to_merge_dict.keys())

# merged_df['date'] = pd.to_datetime(merged_df['date'], unit='ms')
# merged_df
merged_df
```

```python
# set a new df with dates as axis first convert date to datetime
date_index_merge = merged_df
date_index_merge.iloc[:, [0][0]] = pd.to_datetime((date_index_merge.iloc[:, [0][0]]), unit='ms')
date_index_merge
```

```python
# drop na
date_index_merge = date_index_merge.dropna()
date_index_merge
```

```python
# set index to date
date_index_merge.set_index(date_index_merge.iloc[:, [0][0]], drop=True)
```
